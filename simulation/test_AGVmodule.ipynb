```python
optimization = Optimization(sim,client)
object_list = []
position_list = []
for i in range(6):
    object = sim.getObject(f"/OmniPlatform[{i}]")
    object_list.append(object)
    position = sim.getObjectPosition(object)
    position_list.append(position)
print(position_list)
for i in range(6):
    if i == 0:
        sim.setObjectPosition(object_list[i],position_list[3])
    elif i == 1:
        sim.setObjectPosition(object_list[i],position_list[2])
    elif i == 2 or 3:
        sim.setObjectPosition(object_list[i],position_list[i-2])
    elif i == 4 :
        sim.setObjectPosition(object_list[i],position_list[i+1])
    elif i == 5:
        sim.setObjectPosition(object_list[i],position_list[i-1])        
```
